```python
def group_and_summarize_dataframe (df, grouping, stats):      
    # Create an empty dataframe to store the results.
    results_df = pd.DataFrame()
    grouped_df = df.groupby(grouping)
    for name, column, agg_func in stats:
        results_df[name] = grouped_df[column].agg(agg_func)
        
    return results_df.reset_index()
```

```python
xerxes_df = tows_df.loc[tows_df['emergency']=='Xerxes']
```

```python
xerxes_df.head()
```

```python
xerxes_df.shape
```

```python
group_and_summarize_dataframe(xerxes_df, ['day'], [('tows', 'day', 'count')])
```

```python
weekday_dict = {0: "Monday", 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'}
```

```python
xerxes_df.head()
```

```python
xerxes_df['weekday'] = xerxes_df['date'].apply(parse).apply(datetime.datetime.weekday).map(weekday_dict)
```

```python
xerxes_df.head()
```

```python
group_and_summarize_dataframe(xerxes_df, ['weekday'], [('tickets', 'weekday', 'count')])
    
```
